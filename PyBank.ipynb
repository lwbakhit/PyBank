```python
import pandas as pd
from pathlib import Path
import csv
import os
import statistics
import datetime
import numpy as np
```

```python
budget_data_path = Path('../Resources/budget_data.csv')

df = pd.read_csv(budget_data_path) 
df.head()
```

```python
# Read in data
budget_data = pd.read_csv(budget_data_path, parse_dates=True, infer_datetime_format=True)

#budget_data

pd.set_option("display.max_rows", None, "display.max_columns", None)

budget_data
```

```python
#check column names
budget_data.columns
```

```python
#Change column names 

budget_data = budget_data.rename(columns={
    "Date": "months",
    "Profit/Losses": "revenue"
})

budget_data.head()
```

```python
budget_data["revenue"].pct_change().dropna().sum()/85
```

```python
# import numpy as 
budget_data["change"] = np.nan
budget_data.head()
```

```python
for index, row in budget_data.iterrows():
    
    if index>0:
       budget_data.iloc[index, 2] = (row["revenue"] - budget_data.iloc[index-1, 1])
```

```python
budget_data.head()
```

```python
budget_data2 = budget_data.dropna()
budget_data2.mean()
```

```python
# The total number of months included in the dataset.
total_months = df['Date'].count()
total_months
```

```python
# The net total amount of Profit/Losses over the entire period.
profits = df['Profit/Losses'].sum()
profits
```

```python
# Average change in revenue
avg_change = round(budget_data2["change"].sum() / budget_data2["change"].count(), 2)
avg_change
```

```python
# Greatest increase in revenue change
max_val = budget_data2[budget_data2["change"]==budget_data2["change"].max()].reset_index()
maxmonth = max_val.iloc[0, 1]
maxval = max_val.iloc[0, 3]
print(max_val)
```

```python
# greatest decrease in revenue change
min_val = budget_data2[budget_data2["change"]==budget_data2["change"].min()].reset_index()
minmonth = min_val.iloc[0, 1]
minval = min_val.iloc[0, 3]
print(min_val)
```

```python
output = ""
output += f"  Financial Analysis\n"
output += f"  ----------------------------\n"
output += f"  Total Months: {budget_data.count()[0]}\n"
output += f"  Total: ${budget_data['revenue'].sum()}\n"
output += f"  Average  Change: ${avg_change}\n"
output += f"  Greatest Increase in Profits: {maxmonth} (${maxval})\n"
output += f"  Greatest Decrease in Profits: {minmonth} (${minval})\n"
output
```
